```python
BRANDS = ['ford', 'chevrolet', 'plymouth', 'dodge', 'datsun', 'amc', 'buick', 'pontiac', 'volkswagen']
df['brand'] = ''
```

```python
import re

def match_brand(name):
    for b in BRANDS:
        if b in re.findall(r'\w+', name):
            return b
    return 'other'

df['brand'] = df['car name'].apply(match_brand)
```
