```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_table('data/facemasks.tsv').drop_duplicates()
df.set_index('id', verify_integrity=True, inplace=True)
df.head()
```

```python
r = df.groupby('brand')['ratingValue'].mean()
```

```python
s = df.groupby('brand')['product_price'].sum()
```

```python
x = pd.DataFrame.from_dict({'rating': r, 'sales': s})
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.regplot(x='rating', y='sales', data=x)
ax.set_xlabel('Average Product Rating', fontweight='heavy')
ax.set_ylabel('Total Sales (in AU$)', fontweight='heavy')

ax.text(44, 1000, 'Dr Puri')
ax.text(46.071429, 1239.84, 'Hwipure')
ax.text(43.348214, 15877.12, 'Kitsch')
ax.text(45.325670, 17516.20, 'Kosette')
ax.text(46.179775, 609.65, 'Lozperi')
ax.text(46.498516, 5172.95, 'Zidian')
ax.text(37.333333, 4205.20, 'Luseta Beauty')
```
